### The QQ Plot

```python
import scipy.stats
import pylab
```

```python
# Method 1 -> using pylab to show the QQ Plot
scipy.stats.probplot(df_comp.spx,plot=pylab) #1st parameter -> series, 2nd -> what we want to plot
pylab.show()
```

```python
# Method 2 
# using pyplot (plt) to generate QQ plot showing if the data is normally distributed!!
# Mostly this is used now!!
import matplotlib.pyplot as plt
import scipy.stats as stats 

stats.probplot(df_comp.spx, plot=plt)
plt.show()
```
