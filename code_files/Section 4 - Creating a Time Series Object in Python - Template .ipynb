### Handling Missing Values

```python
# To check if the missing values are there in the dataset , we already used .isna().sum() to find the total missing values
df_comp.isna().sum()

# we see that there are 8 missing values to each column!
```

```python
# To handle these missing values, we can fill them using a method .ffill()' and 'bfill()' / or any other value'
df_comp.spx = df_comp.spx.ffill()

 # This code filled the missing values by the one previous day value !!
 # if bfill was used, then it would have filled the nan value by the forward day value
 # we can also fill by using any expression like finding mean, median, mode etc.
```

```python
df_comp.isna().sum() # This code filled the missing values by the one previous day value !!
```

```python
df_comp.ftse = df_comp.ftse.bfill() # backward fill -> fills the nan by one ahead data
df_comp.dax = df_comp.dax.fillna(df_comp.dax.mean()) # mean fill -> fills the nan by the columns mean value (NOT GOOD IDEA FOR TIMEDATA)
```

### Simplifying the Dataset
